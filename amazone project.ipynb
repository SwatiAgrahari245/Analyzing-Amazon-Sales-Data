```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
dataset = pd.read_csv('Amazon Sales data.csv')
```

```python
dataset.head(20)
```

Analysis Missing Values

```python
dataset.isnull().sum()
```

```python
dataset.shape
```

Analysis Order Priority

```python
dataset['Order Priority'].value_counts()
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x="Order Priority",data=dataset)
plt.title("Order Priority")
plt.savefig("Order Priority.jpg")
plt.show()
```

Analysis Sales Channel

```python
dataset['Sales Channel'].value_counts()
```

```python
x=dataset['Sales Channel'].value_counts().index
y=dataset['Sales Channel'].value_counts().values
```

```python
plt.figure(figsize=(5,4))
plt.pie(y,labels=x,startangle=50,autopct="%0.2f%%")
plt.title("Sales Channel")
plt.legend()
plt.savefig("Sales Channel.jpg")
plt.show()
```

Analysis Item Type

```python
dataset['Item Type'].value_counts()
```

```python
plt.figure(figsize=(15,6))
sns.countplot(x="Item Type",data=dataset)
plt.title("Item Type")
plt.savefig("Item Type.jpg")
plt.show()
```

Analysis Item Type In Offline

```python
plt.figure(figsize=(10,5))
sns.countplot(x="Sales Channel",data=dataset[dataset["Sales Channel"]=="Offline"],hue="Item Type")
plt.title("Offline Item Type Sales")
plt.savefig("Offline Item Type.jpg")
plt.show()
```

Analysis Item Type In Online

```python
plt.figure(figsize=(10,5))
sns.countplot(x="Sales Channel",data=dataset[dataset["Sales Channel"]=="Online"],hue="Item Type")
plt.savefig("Online Item Type Sales.jpg")
plt.title("Online Item Type Sales")
plt.show()
```

Analysis Order Year

```python
dataset["Order year"]=dataset['Order Date'].dt.year
```

```python
plt.figure(figsize=(7,4))
sns.countplot(x="Order year",data=dataset)
plt.savefig("Order year.jpg")
plt.show()
```
